데이터 프레임은 데이터를 다룰 때 가장 많이 사용하는 데이터 형태로, 행과 열로 구성된 사각형 모양의 표처럼 생겼다. 어떤 기업에서 연봉에 영향을 미치는 요인을 파악하기 위해 직원을 대상으로 조사했다고 가정해 본다. 

|성별|연령|학점|연봉|
|----|----|----|----|
|남자|26|3.8|2700만원|
|여자|42|4.2|4000만원|
|남자|35|2.6|3500만원|

# '열'은 속성이다.
세로로 나열되는 열은 속성을 나타낸다. 열은 컬럼 또는 변수라고 불린다.

# '행'은 한 사람의 정보다.
가로로 나열되는 행은 각 사람의 정보를 나타낸다. 행은 로 또는 케이스라고 불린다.

# 한 사람의 정보는 가로 한 줄에 나열된다.
앞의 표를 보면 한 사람의 정보는 가로 한 줄에 나열되어 있다. 첫 번재 행을 보면 이 사람은 25세 남자이고, 졸업 학점이 3.8이며, 연봉은 2700만원이라는 걸 알 수 있다. 이처럼 한 명의 데이터느 하나의 행으로 구성된다.

# 하나의 단위가 하나의 행이 된다.
행이 반드시 사람이어야 하는 건 아니다. 무엇이든 하나의 단위가 하나의 행으로 구성될 수 있다. 예를 들어 하나의 도시, 하나의 거래 내역, 하나의 웹 사이트 접속 기록도 하나의 해ㅔㅇ이 될 수 있다. 어떤 데이터가 도시 단위로 되어있고 도시 30개의 정보를 담고 있다면 ,이 데이터는 30행으로 구성된다.

# 데이터가 크다 = 행이 많다 또는 열이 많다.
데이터가 크다는 말은 행이 많다는 의미일 수도 있고, 열이 많다는 의미일 수도 있다. 여기서 문제를 주자면 열이 많은 것이 중요할까? 행이 많은 것이 중요한지 물을 것이다.

정답은 열이 많은 것이 더 중요하다.

100명의 데이터를 분석하다가 데이터가 늘어나 10만 명의 데이터를 분ㅅ헉하게 된다면 데이터 분석가는 컴퓨터가 느려지는 문제를 겪게 된다. 이 문제를 해결하려면 성능 좋은 장비를 구축하는 것이다. 메모리와 CPU를 늘려보고, 그래도 힘들다면 분산 처리 시스템을 구축하거나 클라우드 서비스를 이용하면 된다. 행이 늘어나더라도 분석 기술 면에서는 별다른 차이가 생기지 않는다. 100명의 데이터로 구하든, 10만 명의 데이터로 평균을 구하든 분석 방법이 같다면 데이터 분석에 들이는 노력은 달라지지 않는다.

반면에, 열이 늘어나면 변수를 조합할 수 있는 경우의 수가 늘어난다. 예를 들어 학점과 연봉의 관계는 전공에 따라 다른 양상으로 나타날 수 있다. 여기에 출신지, 자격증, 성별, 직업군 등 수십 가지 변수가 추가되면 경우의 수가 기하급수로 늘어난다. 변수가 늘어나면 단순한 분석 방법으로는 해결하기 힘드므로 여러 변수의 영향을 도잇에 고려하는 복잡한 분석 방법을 활용할 필요성이 생긴다. 머신러닝이 주목받는 이유가 이 때문이다. 이처럼 변수가 많아지면 적용할 분석 기술이 달라지기 때문에 분석가는 전보다 더 노력을 기울여야 한다. 따라서 데이터 분석에는 데이터의 양을 의미하는 행보다 데이터의 다양성을 의미하는 열이 많은 것이 더 중요하다.

## 데이터가 크다

행이 많다 -> 컴퓨터가 느려짐 -> 고사양 장비 구축

열이 많다 -> 분석 방법의 한계 -> 고급 분석 방법
