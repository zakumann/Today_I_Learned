# 패키지 로드하기

패키지의 함수를 사용하려면 우선 패키지를 로드해야 한다. 패키지를 로드하려면 import 뒤에 패키지 이름을 입력하고 실행하면 된다.

```python
import seaborn
```

# 패키지 함수 사용하기

패키지를 로드하고 나면 패키지에 들어 있는 다양한 함수를 이용할 수 있다.seaborn 패키지의 countplot() 함수를 이용해 빈도 막대 그래프를 만들어 본다면, 빈도 막대 그래프는 값의 개수(빈도)를 막대의 길이로 표현한다. 먼저 여러 개의 문자로 구성된 변수 var을 만들어 본다.

```python
var = ['a','a','b','c']
var
```

```python
# countplot()을 이용해서 빈도 막대 그래프를 만들기
seaborn.countplot(x = var)
```

# 패키지 약어

패키지 함수를 이용하려면 함수 이름 앞에 패키지 이름을 항상 입력해야 해서 번거롭다. 패키지 약어를 지정해두면 함수를 사용할 때 피키지 이름 대신 약어를 활용해 코드를 짧게 줄일 수 있다. 패키지 약어를 지정하려면 패키지를 로드하는 코드 뒤에 as와 약어를 입력하면 된다.

```python
# 패키지 약어
import seaborn as sns
```

```python
sns.countplot(x = var)
```

# seaborn의 load_dataset()로 titanic 데이터 불러오기

dataload_dataset()에 'titanic'을 입력해 titanic 데이터를 불러온 다음 df에 할당하기

```python
df = sns.load_dataset('titanic')
df
```

# 함수의 다양한 기능 이용하기 

앞에서 countplot()으로 빈도 막대 그래프를 만들 때 x = var을 입력해 x 축을 var의 값으로 구성하도록 설정했다. countplot()에 입력한 x는 빈도 막대 그래프의 x축을 정하는 기능을 한다. 이처럼 함수의 옵션을 설정하는 명령어를 **파라미터** 또는 **매개변수**라고 한다.

함수에는 저마다 기능이 다른 파라미터가 들어 있고, 이를 조정해서 우너한느 결과를 얻을 수 있다. 따라서 어떤 파라미터가 있는지 알고 있어야 함수를 자유자재로 다룰 수 있다.

```python
# 탑승객의 설별 나타내기
sns.countplot(data = df, x = 'sex')
```

다음 코드에서 class는 선실 등급, alive는 생존 여부를 나타낸 변수이다. hue는 변수의 항목별로 막대의 색을 다르게 표현하는 파라미터이다.

```python
sns.countplot(data = df, x = 'class') # x축 class
sns.countplot(data = df, x = 'class', hue = 'alive') # y축 class, alive별 색 표현
sns.countplot(data = df, y = 'class', hue = 'alive') # y축 class, alive 별 색 표현
```

# 모듈 알아 보기

어떤 패키지는 함수가 굉장히 많기 때문에 비슷한 함수끼리 묶어 몇 개의 모듈로 나뉘어져 있다. 패키지라는 큰 꾸러미에 비슷한 함수들을 넣어둔 작은 꾸러미들이 들어 있다고 생각하면 된다.

머신러닝 모델을 만들 때 사용하는 sklearn 패키지를 예로 들어 모듈을 사용하는 방법을 알아보면, sklearn 패키지에는 metrics, tree, model_selection 등 여러 모듈이 들어 있다. metrics 모듈에는 머신러닝 모델의 예측이 얼마나 정확한지 성능을 평가할 때 사용하는 accuracy_score() 함수가 들어 있다. accuracy_score를 사용하려면 먼저 metrics 모듈을 로드해야 한다. 패키지의 모듈을 불러오려면 import sklearn.metrics처럼 패키지를 로드하는 코드 뒤에 점을 찍고 모듈 이름을 입력하면 된다.

```python
# sklearn 패키지의 metrics 모듈 로드하기
import sklearn.metrics
```

모듈에 들어 있는 함수를 사용하려면 패키지명.모듈명.함수명()을 입력하면 된다.

```python
# sklearn 패키지 metrics 모듈의 accuracy_Score() 사용하기
sklearn.metrics.accuracy_score()
```

# 모듈명.함수명()으로 함수 사용하기

모듈의 함수를 사용할 때 매번 패키지명을 입력하는 게 번거롭다면 from 패키지명 import 모듈명으로 모듈을 로드하면 된다. 모듈을 로드하면 함수를 사용할 때 모듈명. 함수명()만 입력하면 된다.

```python
# sklearn 패키지의 metrics 모듈 로드하기
from sklearn import metrics
metrics.accuracy_score()
```
